```python
with open('train.jw', 'r', encoding='utf-8') as jawi_file:
    jawi_content = jawi_file.read()
    jawi_lines = jawi_content.splitlines()

# Initialize a variable to keep track of the word count
total_words_jawi = 0

# Iterate over each line in the list
for line in jawi_lines:
    # Split the line into words
    words = line.split()
    # Increment the word count by the number of words in the line
    total_words_jawi += len(words)

# Print the total word count
print("Total jawi words:", total_words_jawi)
```

```python
with open('train.jw', 'r', encoding='utf-8') as jawi_file:
    jawi_content = jawi_file.read()
    jawi_lines = jawi_content.splitlines()

# Initialize a list to store unique words
total_words_jawi = []

# Iterate over each line in the list
for line in jawi_lines:
    # Split the line into words
    words = line.split()
    # Iterate over each word in the line
    for word in words:
        # Add the word to the list of unique words if it's not already present
        if word not in total_words_jawi:
            total_words_jawi.append(word)

# Print the unique words
print("Unique jawi words:")
for word in total_words_jawi:
    print(word)

# Count the total unique words
total_unique_words_rumi = len(total_words_jawi)
print("Total unique jawi words:", total_unique_words_rumi)
```
